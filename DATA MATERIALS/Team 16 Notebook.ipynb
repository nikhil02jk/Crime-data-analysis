DATA STANDARDIZATION

```python
print(df1['VICTIM_GENDER'].value_counts())
```

```python
# Define a function to clean and standardize the 'VICTIM_GENDER' values
def clean_victim_gender(gender):
    # Trim whitespace and convert to uppercase
    gender = str(gender).strip().upper()

    # Apply the rules to convert the gender values
    if gender in ['MALE', 'M - MALE']:
        return 'MALE'
    elif gender in ['FEMALE', 'F - FEMALE']:
        return 'FEMALE'
    elif gender == 'NON-PERSON (BUSINESS':
        return 'UNKNOWN'
    else:
        return 'UNKNOWN'

# Apply the function to the 'VICTIM_GENDER' column
df2['VICTIM_GENDER'] = df2['VICTIM_GENDER'].apply(clean_victim_gender)

# Display the cleaned column to verify
print(df2['VICTIM_GENDER'].value_counts())
```

```python
print(df1['SUSPECT_GENDER'].value_counts())
```

```python
# Define a function to clean and standardize the 'VICTIM_GENDER' values
def clean_suspect_gender(gender):
    # Trim whitespace and convert to uppercase
    gender = str(gender).strip().upper()

    # Apply the rules to convert the gender values
    if gender in ['MALE', 'M - MALE']:
        return 'MALE'
    elif gender in ['FEMALE', 'F - FEMALE']:
        return 'FEMALE'
    elif gender == 'NON-PERSON (BUSINESS':
        return 'UNKNOWN'
    else:
        return 'UNKNOWN'

# Apply the function to the 'VICTIM_GENDER' column
df2['SUSPECT_GENDER'] = df2['SUSPECT_GENDER'].apply(clean_suspect_gender)

# Display the cleaned column to verify
print(df2['SUSPECT_GENDER'].value_counts())
```

OUTLIER DETECTION

```python
import pandas as pd

# Assuming df is your DataFrame
# Columns to check for outliers
columns = [ 'LONGITUDE', 'LATITUDE']

# Function to detect outliers using IQR
def detect_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Detect outliers in each numerical column
outliers = {}
for col in columns:
    outliers[col] = detect_outliers_iqr(df3, col)

# Check the outliers for each column
for col, outliers_df in outliers.items():
    print(f"Outliers for {col}:")
    print(outliers_df[[col]], "\n")
```

We have found the outliers for the data for 1092 rows in latitude and 84 rows in longitiude, but we have to take the decison of wether to remove this data or not

```python
# Assuming longitude and latitude are in separate columns
df_new_filtered = df3[(df3['LONGITUDE'] < -180) | (df3['LONGITUDE'] > 180) |
                       (df3['LATITUDE'] < -90) | (df3['LATITUDE'] > 90)]

print(f"Invalid Coordinates:\n{df_new_filtered[['LONGITUDE', 'LATITUDE']]}")
```

we can see that the latitide and longituide values are in the range of the correct value, which means it is valid  , hence will not be deleting this

```python
import pandas as pd

# Assuming df is your DataFrame
# Columns to check for outliers
columns = ['TOTALNUMBERVICTIMS', 'TOTALSUSPECTS']

# Function to detect outliers using IQR
def detect_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Convert columns to numeric, coerce errors to NaN (to handle invalid data)
df1[columns] = df1[columns].apply(pd.to_numeric, errors='coerce')

# Now run your outlier detection after ensuring the columns are numeric
outliers = {}
for col in columns:
    outliers[col] = detect_outliers_iqr(df1, col)

# Print distinct outliers for each column
for col, outliers_df in outliers.items():
    print(f"Distinct Outliers for {col}:")
    print(outliers_df[col].unique(), "\n")

```

We can see that there are outliers in the data, in the column TOTALNUMBERVICTIMS and TOTALSUSPECTS , but we also can observe that the data is valid, because these are in the range of true values , hence we cannot dump these data as well
